```python
import numpy as np

# Users
friends = ["Ram", "Sita", "Arun", "Bala"]

# Expense matrix
expense_matrix = np.zeros((len(friends), len(friends)))

# Transaction history
history = []

# Add expense
def add_expense():
    payer = input("Enter payer name: ")
    if payer not in friends:
        print("Invalid user")
        return

    amount = float(input("Enter amount: "))
    beneficiaries = input("Enter beneficiaries (comma separated): ").split(",")

    beneficiaries = [b.strip() for b in beneficiaries if b.strip() in friends]

    if not beneficiaries:
        print("No valid beneficiaries")
        return

    split_type = input("Split type (equal/custom): ")

    payer_idx = friends.index(payer)

    if split_type == "equal":
        share = amount / len(beneficiaries)
        for b in beneficiaries:
            b_idx = friends.index(b)
            expense_matrix[payer_idx][b_idx] += share

    elif split_type == "custom":
        print("Enter amounts for each:")
        for b in beneficiaries:
            amt = float(input(f"{b}: "))
            b_idx = friends.index(b)
            expense_matrix[payer_idx][b_idx] += amt

    history.append(f"{payer} paid {amount} for {', '.join(beneficiaries)}")
    print("Expense added successfully!")

# Show balances
def show_balances():
    total_paid = np.sum(expense_matrix, axis=1)
    total_owed = np.sum(expense_matrix, axis=0)
    net_balance = total_paid - total_owed

    print("\n--- Balance Summary ---")
    for i in range(len(friends)):
        print(f"{friends[i]}: {round(net_balance[i], 2)}")

# Show history
def show_history():
    print("\n--- Transaction History ---")
    for h in history:
        print(h)

# Menu
def main():
    while True:
        print("\n1. Add Expense")
        print("2. Show Balances")
        print("3. Show History")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            show_balances()
        elif choice == "3":
            show_history()
        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice")

if __name__ == "__main__":
    main()
```
